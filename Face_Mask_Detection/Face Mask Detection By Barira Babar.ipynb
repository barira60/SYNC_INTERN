```python
import os
from PIL import Image
import numpy as np

with_mask_path = 'C:\\Users\\Admin\\dataset\\with mask'
without_mask_path = 'C:\\Users\\Admin\\dataset\\without mask'

data = []

# List image files in with_mask directory
with_mask_files = os.listdir(with_mask_path)

for img_file in with_mask_files:
    image = Image.open(os.path.join(with_mask_path, img_file))
    image = image.resize((128, 128))
    image = image.convert('RGB')
    image = np.array(image)
    data.append(image)

# List image files in without_mask directory
without_mask_files = os.listdir(without_mask_path)

for img_file in without_mask_files:
    image = Image.open(os.path.join(without_mask_path, img_file))
    image = image.resize((128, 128))
    image = image.convert('RGB')
    image = np.array(image)
    data.append(image)
```

```python
from sklearn.model_selection import train_test_split


# Determine the size of the smaller dataset (use min() to avoid errors)
min_num_samples = min(X.shape[0], Y.shape[0])

# Split the data based on the size of the smaller dataset
X_train, X_test, Y_train, Y_test = train_test_split(X[:min_num_samples], Y[:min_num_samples], test_size=0.2, random_state=2)

# Now, X_train, X_test, Y_train, and Y_test have the same number of samples
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
X_train_scaled = X_train/255

X_test_scaled = X_test/255
```

```python
X_train[0]
```

```python
X_train_scaled[0]
```

```python
import keras
```

```python
num_of_classes = 2

model = keras.Sequential()

model.add(keras.layers.Conv2D(32, kernel_size=(3,3), activation='relu', input_shape=(128,128,3)))
model.add(keras.layers.MaxPooling2D(pool_size=(2,2)))


model.add(keras.layers.Conv2D(64, kernel_size=(3,3), activation='relu'))
model.add(keras.layers.MaxPooling2D(pool_size=(2,2)))

model.add(keras.layers.Flatten())

model.add(keras.layers.Dense(128, activation='relu'))
model.add(keras.layers.Dropout(0.5))

model.add(keras.layers.Dense(64, activation='relu'))
model.add(keras.layers.Dropout(0.5))


model.add(keras.layers.Dense(num_of_classes, activation='sigmoid'))
```

```python
# compile the neural network
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['acc'])
```

```python
# training the neural network
history = model.fit(X_train_scaled,Y_train, validation_split=0.1, epochs=5)
```

```python
loss, accuracy = model.evaluate(X_test_scaled, Y_test) 
print('Test Accuracy =', accuracy)
```

```python
h = history

# plot the loss value
plt.plot(h.history['loss'], label='train loss')
plt.plot(h.history['val_loss'], label='validation loss')
plt.legend()
plt.show()

# plot the accuracy value
plt.plot(h.history['acc'], label='train accuracy')
plt.plot(h.history['val_acc'], label='validation accuracy')
plt.legend()
plt.show()
```
